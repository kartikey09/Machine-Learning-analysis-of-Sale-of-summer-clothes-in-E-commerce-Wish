```python
data.drop(['title', 'title_orig', 'price', 'currency_buyer', 'badges_count',
       'badge_local_product', 'badge_product_quality',
       'badge_fast_shipping', 'product_color',
       'product_variation_size_id', 
       'shipping_option_name', 'tags',
       'shipping_is_express', 'countries_shipped_to', 'inventory_total',
       'has_urgency_banner', 'urgency_text', 'origin_country',
       'merchant_title', 'merchant_name', 'merchant_info_subtitle',
       'merchant_rating_count', 'merchant_rating', 'merchant_id',
       'merchant_has_profile_picture', 'merchant_profile_picture',
       'product_url', 'product_picture', 'product_id', 'theme',
       'crawl_month'], axis = 1, inplace = True)
data.head()
```

```python
data.fillna(0, inplace = True)
```

```python
X = data.drop(['units_sold'], axis = 1).values
y = data['units_sold'].values
```

```python
X.shape, y.shape
```

```python
y.reshape([1573,1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
classifiers = [AdaBoostRegressor(), 
               GradientBoostingRegressor(), 
               RandomForestRegressor(), 
               DecisionTreeRegressor(), 
               KNeighborsRegressor(3),
               XGBRegressor(),
               LinearRegression()]

log_cols = ['classifier', 'accuracy']
log = pd.DataFrame(columns = log_cols)
acc_dict = {}

for clf in classifiers:
    name = clf.__class__.__name__
    clf.fit(X_train, y_train)
    prediction = clf.predict(X_test)
    acc = r2_score(y_test, prediction)
    if name in acc_dict:
        acc_dict[name] += acc
    else:
        acc_dict[name] = acc
acc_dict
```

## Hyperparameter tuning for gradient boost

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
```

```python
# reg = GradientBoostingRegressor()
# reg.fit(X_train, y_train)
# pred = reg.predict(X_test)
# accuracy = r2_score(y_test, pred)
```

```python
parameters = {'loss': ['ls'], 
              'learning_rate': [0.2, 0.4, 0.6, 0.8],
              'n_estimators': [50, 100, 125,],
              'max_depth': [3,8,16,32]}

```

```python
regmod = GradientBoostingRegressor()
keep_accuracy = []
loss1 = 0
learning_rate1 = 0
n_estimators1 = 0
max_depth1 = 0
max = 0
for i in parameters['loss']:
    for j in parameters['learning_rate']:
        for k in parameters['n_estimators']:
            for l in parameters['max_depth']:
                GradientBoostingRegressor(loss = i, learning_rate = j, n_estimators = k, max_depth = l)
                regmod.fit(X_train, y_train)
                pred = regmod.predict(X_test)
                accuracy = r2_score(y_test, pred)
                keep_accuracy.append(accuracy)
                if accuracy > max:
                    loss1 = i
                    learning_rate1 = j
                    n_estimators1 = k
                    max_depth1 = l
                    max = accuracy

```

```python
print('accuracy = ', max, 'loss = ', loss1, 'learning rate = ',learning_rate1, 'n_estimators = ', n_estimators1, 'max_depth = ', max_depth1 )
```
